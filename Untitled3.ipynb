**Vectorization using TF-IDF**

```python
#we'll find out the Term Frequency-Inverse Document Frequency (TF-IDF) vectors for each overview.
#It is used to calculate the relative frequency of a word in a document
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# minimum document freq =3
# we remove the unnecessary words and characters
tfidf = TfidfVectorizer(min_df=3, max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')
# minimum document freq =3
```

```python
# Fitting the TF-IDF on the 'overview' text
tfidf_matrix = tfidf.fit_transform(df['overview'])
```

```python
tfidf_matrix.shape
```

```python
# the above output-tells us that over 10K words are used to describe 4799 movies 
```

**finding the similarity score between two movies**

```python
#Kernels are measures of similarity, i.e. s(a, b) > s(a, c) if objects a and b are considered “more similar” than objects a and c
#The function sigmoid_kernel computes the sigmoid kernel between two vectors

from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfidf_matrix, tfidf_matrix)
```

```python
# Reverse mapping of indices and movie titles - here we fetch the index from the movie titles
indices = pd.Series(df.index, index=df['original_title']).drop_duplicates()
```

**Defining the get_recommendation function**

```python
indices['The Dark Knight Rises']
```

```python
sig[3] #gives the similarity score of The dark knight rises with all other movies
```

```python
list(enumerate(sig[3]))
```

```python
def get_recommendations(title, sig=sig):
    # step1 Get the index of the movie that matches with the movie name
    idx = indices[title]

    # step2 Get the sigmoid similarity scores of all movies with that particular movie
    scores = list(enumerate(sig[idx]))

    #step3 Sort the movies based on the similarity scores
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    #step4 Get the scores of the 5 most similar movies
    scores = scores[1:6]

    #step5 Get the movie indices
    movie_indices = [i[0] for i in scores]

    # Return the top 5 most similar movies
    return df['original_title'].iloc[movie_indices]
```

```python
get_recommendations('Titanic')
```

**Convertion of stringified lists to normal python lists**

```python
from ast import literal_eval as le
example= le(df['genres'][0])
print(example)
print(type(example))
```

```python
# applying literal_eval to the following features
features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df[feature] = df[feature].apply(le)
```

```python
#The next step is to extract the required info. from each feature.
```

```python
# If the director's name is present inside the crew feature return it else return NaN
def get_director(x):
    for i in x:
        if i.get('job') == 'Director':
            return i['name']
    return np.nan
```

```python
#Check if more than 3 elements exist. If yes, return only first three. If no, return entire list.

def get_list(x):
    if isinstance(x, list): #isinstance(object, type) it checks if the object is of the given type
        names = [i['name'] for i in x]
        
        if len(names) > 3:
            names = names[:3]
        return names

    #Return empty list in case of missing/malformed data
    return []
```

```python
# Define new director, cast, genres and keywords features that are in a correct form.
df['director'] = df['crew'].apply(get_director)

features = ['cast', 'keywords', 'genres']
for feature in features:
    df[feature] = df[feature].apply(get_list)
```

```python
df[['original_title', 'cast', 'director', 'keywords', 'genres']].head()
```

```python
# since there can be a lot of similarities in the first names of the director/cast , we need to convert them into lowercase
# and remove the spaces between them so that they can be distinguished on the basis of their surname
```

```python
def clean(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, if it is Nan return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
#cleaning the data
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    df[feature] = df[feature].apply(clean)
```

```python
# next step is to create a metadata soup", which is a string that contains all the metadata that we want to feed 
#to our vectorizer
```

```python
def metadata_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df['soup'] = df.apply(metadata_soup, axis=1)
```

**Vectorization using CountVectorizer**

```python
#CountVectorizer counts the frequency of all words in our corpus, sorts them and grabs the most recurring features.
# In TF-IDF common words are penalized . So we have to use count verctorizer as we are prioritising the occurance 
# of the particular director/cast in multiple movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(strip_accents='unicode', 
                        analyzer='word',
                        token_pattern=r'\w{1,}',
                        ngram_range=(1, 3),
                        stop_words='english')
count_matrix = count.fit_transform(df['soup'])
```

```python
# Computing the sigmoid Similarity matrix based on the count_matrix
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(count_matrix, count_matrix)
```

```python
# Resetting the index of the df and constructing the reverse mapping
df = df.reset_index()
indices = pd.Series(df.index, index=df['original_title'])
```
